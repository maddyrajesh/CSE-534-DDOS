```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Add the 'attack_prediction' column to both dataframes
df_legitimate['attack_prediction'] = 0
df_attack['attack_prediction'] = 1

# Concatenate the two dataframes
df_combined = pd.concat([df_legitimate, df_attack], ignore_index=True)

# Display the combined dataframe
print(df_combined.head())

# Compute the confusion matrix
y_true = df_combined['attack']
y_pred = df_combined['attack_prediction']
cm = confusion_matrix(y_true, y_pred)

print("Confusion matrix:")
print(cm)
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Add the 'attack_prediction' column to both dataframes
df_legitimate['attack_prediction'] = 0
df_attack['attack_prediction'] = 1

# Concatenate the two dataframes
df_combined = pd.concat([df_legitimate, df_attack], ignore_index=True)

# Compute the confusion matrix
y_true = df_combined['attack']
y_pred = df_combined['attack_prediction']
cm = confusion_matrix(y_true, y_pred)

# Plot the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.4)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Legitimate', 'Attack'], yticklabels=['Legitimate', 'Attack'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Add the 'attack_prediction' column to both dataframes
df_legitimate['attack_prediction'] = 0
df_attack['attack_prediction'] = 1

# Concatenate the two dataframes
df_combined = pd.concat([df_legitimate, df_attack], ignore_index=True)

# Compute the AUC-ROC score
y_true = df_combined['attack']
y_pred = df_combined['attack_prediction']
auc_roc = roc_auc_score(y_true, y_pred)

# Compute the ROC curve
fpr, tpr, thresholds = roc_curve(y_true, y_pred)

# Plot the ROC curve
plt.plot(fpr, tpr, label=f'AUC = {auc_roc:.4f}')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
