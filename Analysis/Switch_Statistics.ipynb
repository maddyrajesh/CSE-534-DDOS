```python
import csv

# Open the text file for reading
with open('input.txt', 'r') as infile:

    # Open a CSV file for writing
    with open('output.csv', 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        
        # Write the header row to the CSV file
        writer.writerow(['Timestamp', 'Source IP Entropy', 'Source IP EWMA', 'Source IP EWMMD', 'Destination IP Entropy', 'Destination IP EWMA', 'Destination IP EWMMD', 'DDoS Alarm'])

        # Process each line of the input file
        for line in infile:
            # Split the line into fields
            fields = line.strip().split()

            # Write the fields to the CSV file
            writer.writerow(fields)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV into DataFrame
df = pd.read_csv("output.csv")

# Set Timestamp column as index and convert to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"], unit='us')
df.set_index("Timestamp", inplace=True)

# Assuming your DataFrame for CPU and Memory is named 'df_cpu_mem'
# If you're reading from a CSV file, uncomment the line below
df_cpu_mem = pd.read_csv('/content/output_02.csv')

# Convert the 'Timestamp' column to datetime objects
df_cpu_mem['Timestamp'] = pd.to_datetime(df_cpu_mem['Timestamp'])

# Calculate the time offset in seconds
df_cpu_mem['Time_Offset'] = (df_cpu_mem['Timestamp'] - df_cpu_mem['Timestamp'].min()).dt.total_seconds()

# Create a 4x2 grid of subplots
fig, axs = plt.subplots(4, 2, figsize=(16, 24), sharey=False)

# Plot each column from the output.csv DataFrame in a subplot
for i, column_name in enumerate(df.columns):
    row = i // 2
    col = i % 2
    ax = axs[row, col]  # get the subplot axis
    ax.plot((df.index - df.index[0]).total_seconds(), df[column_name])
    ax.set_ylabel(column_name)
    ax.set_xlabel("Time (s)") # add x-axis label to subplot
    
    # modify y-axis of DDoS alarm plot
    if column_name == "DDoS Alarm":
        ax.set_yticks([0, 1])


# Plot CPU and Memory usage on the last two subplots
axs[3, 0].plot(df_cpu_mem['Time_Offset'], df_cpu_mem['%CPU'], label='CPU Usage (%)')
axs[3, 0].set_xlabel('Time (s)')
axs[3, 0].set_ylabel('CPU Usage (%)')
axs[3, 0].legend()

axs[3, 1].plot(df_cpu_mem['Time_Offset'], df_cpu_mem['%MEM'], label='Memory Usage (%)')
axs[3, 1].set_xlabel('Time (s)')
axs[3, 1].set_ylabel('Memory Usage (%)')
axs[3, 1].legend()

# Add padding between subplots
fig.subplots_adjust(wspace=0.5, hspace=0.5)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV into DataFrame
df = pd.read_csv("output.csv")

# Set Timestamp column as index and convert to datetime format
df["Timestamp"] = pd.to_datetime(df["Timestamp"], unit='us')
df.set_index("Timestamp", inplace=True)

# Assuming your DataFrame for CPU and Memory is named 'df_cpu_mem'
# If you're reading from a CSV file, uncomment the line below
df_cpu_mem = pd.read_csv('/content/output_02.csv')

# Convert the 'Timestamp' column to datetime objects
df_cpu_mem['Timestamp'] = pd.to_datetime(df_cpu_mem['Timestamp'])

# Calculate the time offset in seconds
df_cpu_mem['Time_Offset'] = (df_cpu_mem['Timestamp'] - df_cpu_mem['Timestamp'].min()).dt.total_seconds()

# Create a 2x4 grid of subplots
fig, axs = plt.subplots(2, 4, figsize=(24, 12), sharey=False)

# Plot each column from the output.csv DataFrame in a subplot
for i, column_name in enumerate(df.columns[:-1]):  # Exclude the last column (DDoS Alarm)
    row = i // 4
    col = i % 4
    ax = axs[row, col]  # get the subplot axis
    ax.plot((df.index - df.index[0]).total_seconds(), df[column_name])
    ax.set_ylabel(column_name)
    ax.set_xlabel("Time (s)") # add x-axis label to subplot

# Plot CPU and Memory usage on the same subplot
axs[1, 2].plot(df_cpu_mem['Time_Offset'], df_cpu_mem['%CPU'], label='CPU Usage (%)')
axs[1, 2].plot(df_cpu_mem['Time_Offset'], df_cpu_mem['%MEM'], label='Memory Usage (%)')
axs[1, 2].set_xlabel('Time (s)')
axs[1, 2].set_ylabel('Usage (%)')
axs[1, 2].legend()

# Plot DDoS Alarm on the last subplot
axs[1, 3].plot((df.index - df.index[0]).total_seconds(), df['DDoS Alarm'])
axs[1, 3].set_ylabel('DDoS Alarm')
axs[1, 3].set_xlabel("Time (s)") # add x-axis label to subplot
axs[1, 3].set_yticks([0, 1])

# Add padding between subplots
fig.subplots_adjust(wspace=0.2, hspace=0.1)

plt.show()
```
